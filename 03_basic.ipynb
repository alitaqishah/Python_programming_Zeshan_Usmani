# Basic assignment statements
- assignment statement python main basic importance rakhti hai . hum in ka through aik expression sa banne wale object ko save krte hai .
- basically assignment statement ka through hum variables banate aur change krte hai .
- ye statments aik expression aur aik variables provide krti hai . jo expresion ki value ko label krne ka liye use ki jati hai .


```python
# compute the value of a block of stock
shares = 150
price = 3 + 5.0/8.0
value = shares * price
print("value",value)
```

kisi bhi arithmetic operator ko assignment operator ka saath mila kr aik "augmented statement" ready ki ja sakti hai 

```python
# Total value of a portfolio made up of two blocks of stock
portfolio = 0
portfolio += 150 * 2 + 1/4.0
portfolio += 75 *1 + 7/8.0

print("portfolio", portfolio)
```

# Variables & their rules
- variable ka start kisi word ya (_) underscore sa hone chahiye.
- variable ka start kisi number ka saath nahi ho sakta.
- variable ka andar sirf alpha numeric aur underscore aa saktay hai 
- variable ka naam case-sensitive hote hai jese ka (age, Age, AGE) ye 3 variables hai 
- variables ko pehlay declare krne ki zaroorat nahi hoti matlab a = 2 , is main variable 'a' hai jis ki value 2 hai.
- aap aik hi line main aik sa ziada variable bhi bana sakte hai
- aap aik hi line main aik sa ziada variable ki value bhi store kr sakte hai 

```python
x, y, z = "apple", "banana", "orange" # aik line main aik sa ziada variable
print(x)
print(y)
print(z)
```

```python
x = y = z = "apple" # aik line main aik sa ziada variable ki value store krna
print(x)
print(y)
print(z)
```
