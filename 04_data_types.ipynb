# 5. Set types ( set, frozenset)
- set aik aese collection of objects hai jo un-organized aur un-indexed hai.
- ye aese data structure hai jo sequence ko store krte hai .
- set ko index krna ya subset krna possible nahi hai .
- set kisi bhi kisam ka data (int, float, strings, bool) store kr saktay hai .
- elements ko set ka andar include ya remove nahi kia ja sakta.
- set sirf unique values store krta hai is main duplicate values nahi aa sakti.
- agar koi value 2 dafa aay tou ye issay sirf 1 dafa ki consider krte hai.
- is pe arithmetic set operation bhi perform kiye ja saktay hai jese ka union , intersection .
- python main set {} curly brackets main likhe jate hai .
- dictionary aur sets {} brackets main likhe jate hai .
- set main aik item ko add krne ka liye add() method use kia jata hai.
- set main aik sa ziada items add krne ka liye update() method use kia jata hai.
- set ki length find krne ka liye len() ka method use kia jata hai.
- set main item ko nikalne ka liye remove() ya discard() ka method use kia jata hai.

```python
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
# add method

thisset = {"apple", "banana", "cherry"}
thisset.add("orange")
print(thisset)
```

```python
# update method

thisset = {"apple", "banana", "cherry"}

thisset.update(["orange", "mango", "grapes"])

print(thisset)
```

```python
# lenght method

thisset = {"apple", "banana", "cherry"}
print(len(thisset))
```

```python
# remove method

thisset = {"apple", "banana", "cherry"}
thisset.remove("banana")

print(thisset)
```

# Python dates
- python main date aur time ka liye aik built-in module datetime use hota hai.
- because date aur time ka format int aur float ka format sa different hota hai is liye issay different tariqa sa handle kia jata hai.


```python
# current time

import datetime
x = datetime.datetime.now()
print(x)
```

```python
# day aur year ka naam display

import datetime
x = datetime.datetime.now()
print(x.year)
print(x.strftime("%A"))

```

```python
# individually day, year, month

import datetime
x = datetime.datetime.now() # current date and time
year = x.strftime(("%Y"))
month = x.strftime(("%m"))
day = x.strftime(("%d"))
print("day: ", day)
time = x.strftime("%H: %M, %S")
print("time: ", time)
date_time = x.strftime("%m/%d/%Y, %H:%M:%S")
print("date and time:", date_time)
```

format string banane ka liye strftime ka method use kia jata hai .

```python
from datetime import datetime
# current date and time

now = datetime.now()
t = now.strftime("%H: %M: %S")
print("time: ", t)
s1 = now.strftime("%m/%d/%Y")
# mm/dd/YY format
print("s1:", s1)

s2 = now.strftime("%d/%m/Y")
# dd/mm/YY format

s3 = now.strftime("%d-%m-%Y")
# dd-mm-YY format

s4 = now.strftime("%a-%b-%y")
print("s4:", s4)

s5 = now.strftime("%A-%B-%Y")
print("s5:", s5)
```
